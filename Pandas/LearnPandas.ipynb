```python
#Making changes to data (New Columns)
df['Count'] = 1
#Rearranging Columns
cols = list(df.columns.values)
df = df [cols[0:4] + [cols[-1]] + cols[5:10]]
df
```

```python
#Filtering Data
df.loc[(df['Type 1'] == 'Poison') & (df['Type 2'] == 'Dark')]
df.loc[(df['Type 1'] == 'Poison') | (df['Type 1'] == 'Dark') & (df['SP Atk'] > 100)]
new_df = df.loc[(df['Type 1'] == 'Poison') | (df['Type 1'] == 'Dark') & (df['SP Atk'] > 100)]
#for new index, use reset_index, and for dropping old index use drop = True, and for self reset use inplce = True
new_df.reset_index(drop = True, inplace = True)
new_df
#Regex Filtering re stands for regular expressions
import re

df.loc[df['Name'].str.contains('Mega')]
#note (~ if for not in pandas)
df.loc[~df['Name'].str.contains('Mega')]

df.loc[df['Type 1'].str.contains('fire|grass',flags = re.I,regex = True)]
df.loc[df['Name'].str.contains('bu[a-z]*',flags = re.I,regex = True)]
df.loc[df['Name'].str.contains('^bu[a-z]*',flags = re.I,regex = True)]
```

```python
#Conditional Changes
df.loc[df['Type 1'] == 'Grass', 'Name'] = 'Graminha'
df
```

```python
#Aggregate Statistics (Groupby)
df.groupby(['Type 1']).mean()
df.groupby(['Type 1']).mean().sort_values('Spd', ascending = False)
df.groupby(['Type 1']).sum()
df.groupby(['Type 1']).count()['Count']
df.groupby(['Type 1', 'Type 2']).count()['Count']
```
